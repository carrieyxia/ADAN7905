## Load

```python
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from sqlalchemy import Date

Base = declarative_base()

class GDP(Base):
    __tablename__ = 'gdp'
    date = Column(Date, primary_key=True)
    gdp = Column(Float)
    rgdp = Column(Float)
    rgdp_per_capita = Column(Float)

class Inflation(Base):
    __tablename__ = 'inflation'
    date = Column(Date, primary_key=True)
    inflation = Column(Float)
    
class DPI(Base):
    __tablename__ = 'dpi'
    date = Column(Date, primary_key=True)
    dpi = Column(Float)
    
class FederalInterestPayment(Base):
    __tablename__ = 'fed_interest_payment'
    date = Column(Date, primary_key=True)
    interest_payment = Column(Float)
    
class CorporateBondIndex(Base):
    __tablename__ = 'cor_bond_index'
    date = Column(Date, primary_key=True)  # Use the 'name' parameter to map to 'DATE' column
    corporate_bond_index = Column(Float)

class RealPersonalExpenditure(Base):
    __tablename__ = 'real_personal_expend'
    date = Column(Date, primary_key=True)
    real_personal_expenditure = Column(Float)

class BusinessExpectations(Base):
    __tablename__ = 'business_expectations'
    date = Column(Date, primary_key=True)
    business_expectations_rev_growth = Column(Float)

class SP100Daily(Base):
    __tablename__ = 'sp100'
    date = Column(Date, primary_key=True)
    open = Column(Float)
    high = Column(Float)
    low = Column(Float)
    close = Column(Float)
    adj_close = Column(Float)
    volume = Column(Float)   

engine = create_engine('postgresql://postgres:17110303Cx@localhost:5432/econ')
Base.metadata.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()

gdp.to_sql('gdp', con=engine, if_exists='replace', index=True, dtype={'date': Date})
dpi.to_sql('dpi', con=engine, if_exists='replace', index=True, dtype={'date': Date})
inflation.to_sql('inflation', con=engine, if_exists='replace', index=True, dtype={'date': Date})
fed_interest_payment.to_sql('fed_interest_payment', con=engine, if_exists='replace', index=True, dtype={'date': Date})
cor_bond_index.to_sql('cor_bond_index', con=engine, if_exists='replace', index=True, dtype={'date': Date})
real_personal_expend.to_sql('real_personal_expend', con=engine, if_exists='replace', index=True, dtype={'date': Date})
business_expectations.to_sql('business_expectations', con=engine, if_exists='replace', index=True, dtype={'date': Date})
sp100_daily.to_sql('sp100', con=engine, if_exists='replace', index=True, dtype={'Date': Date})

session.commit()
```

## Transform

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

# Plot GDP
plt.plot(gdp.index, gdp['gdp'], label='GDP', color='blue')

# Plot S&P 100 Close values
plt.plot(sp100_daily.index, sp100_daily['Close'], label='S&P 100 Close', color='green')

# Customize the plot
plt.title('GDP and S&P 100 Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

### normalize

```python
gdp_resampled = gdp.resample('Q').mean().ffill()
dpi_resampled = dpi.resample('Q').mean().ffill()
inflation_resampled = inflation.resample('Q').mean()
cor_bond_index_resampled = cor_bond_index.resample('Q').mean()
business_expectations_resampled = business_expectations.resample('Q').mean()
sp100_daily_resampled = sp100_daily.resample('Q').mean()
real_personal_expend_resampled = real_personal_expend.resample('Q').mean().ffill()
fed_interest_payment_resampled = fed_interest_payment.resample('Q').mean().ffill()
```

```python
sp100_daily_resampled = sp100_daily_resampled.rename_axis("DATE")
sp100_daily_resampled.head()
```

```python
sp100_combined = pd.concat([sp100_daily_resampled, gdp_resampled, dpi_resampled, inflation_resampled, cor_bond_index_resampled, business_expectations_resampled, real_personal_expend_resampled, fed_interest_payment_resampled], axis=1)
sp100_combined.head()
```

```python
sp100_combined['price'] = sp100_combined[['Open', 'Close', 'High', 'Low', 'Adj Close']].mean(axis=1)
sp100_combined.drop(['Open', 'Close', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
```

### Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = sp100_combined.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
