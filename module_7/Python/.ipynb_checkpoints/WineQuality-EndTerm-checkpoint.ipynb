#### Mising Value Analysis

```python
def missing_data(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100)
    tt = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return tt
```

```python
# Check missing value percentage in dataframe
## 4. <TODO> 
# TODO - 1 lines of code 
print(missing_data(dfWine))
```

#### Create feature matrix

```python
## 6. <TODO> - using wineClass column from dfWine create target varaible dfY
##           - usng remaing columns from dfWine create feature matrix dfX 
# TODO - 2 lines of code 
dfY = dfWine['wineClass']
dfX = dfWine.drop('wineClass', axis=1)
```

```python
dfWine
```

#### Scaling the features

```python
# Scaling The Data

from sklearn.preprocessing import StandardScaler
## 7. <TODO> - use StandardScalar to scale dfX
# TODO - 3 lines of code 

scaler = StandardScaler()
dfX_scaled = scaler.fit_transform(dfX)
dfX =  pd.DataFrame(dfX_scaled, columns=dfX.columns)
```

```python
dfX
```

```python
# To check for outliers, 
## 8. <TODO> - using seaborn package plot a box-whisker plot
## 8. <TODO> - using seaborn package plot a box-whisker plot
# TODO - 3 lines of code  
sns.boxplot(data=dfX, orient='h')
plt.show()
```

#### Train for classification

```python
# Splitting data to train and test

from sklearn.model_selection import train_test_split

#Split the data into test and train 80:20 ratio using tran_test_split
## 9. <TODO> 
X_train, X_test, Y_train, Y_test = train_test_split(dfX_scaled, dfY, test_size=0.2)
```

```python
#Logistic regression

from sklearn.linear_model import LogisticRegression


## 10. <TODO> - using LogisticRegression perfom fit and predict

logreg_model = LogisticRegression()
logreg_model.fit(X_train, Y_train)
Y_pred = logreg_model.predict(X_test)
```

```python
# Generating confusion matrix, accuracy score and classification report to check score of our prediction.

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm_LR=confusion_matrix(Y_test,Y_pred)
print("Confusion Matrix:")
print(cfm_LR)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc_LR=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc_LR)
```
